## 전국 교육대수

```python
edu_uinver =  pd.read_csv('./미니1차/전국교육대수.csv', encoding='euc-kr')
edu_uinver = edu_uinver.drop(0, axis=0)
edu_uinver
```

```python
edu_uinver = edu_uinver.set_index('시도별(1)')
edu_uinver
```

```python
#ecu_uinver.loc['경기',['2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']] = 0
```

```python
# 위에서 구한 각 학교의 수의 총계를 뽑아 하나의 데이터프레임으로 합치기
all_school = pd.concat([eleschool.loc[['총계']],midschool.loc[['총계']],highschool.loc[['총계']],
          uinver.loc[['총계']],uinversity.loc[['총계']],edu_uinver.loc[['총계']]], axis=0)
# all_school = all_school.rename(index={'0':'초등학교','1':'중학교','2':'고등학교',
#                                      '3':'전문대학','4':'대학교','5':'교육대학'})
all_school = all_school.reset_index(drop=False)
all_school
```

```python
# 0번째 행의 특정 컬럼값 변경
all_school.loc[0,'시도별(1)'] = '초등학교'
```

```python
# all_school = all_school.drop('시도별', axis=1)
```

```python
# 특정 행의 특정 컬럼 값 변경
all_school.loc[[1,2,3,4,5],'시도별(1)'] = ['중학교','고등학교','전문대','대학교','교육대']
```

```python
# 변경한 값의 컬럼을 인덱스로 설정
all_school = all_school.set_index('시도별(1)')
all_school
```

```python
# 특정 행을 열방향으로 더하고 '대학교' 행에 대입
all_school.loc['대학교'] = all_school.loc['전문대'] + all_school.loc['대학교'] + all_school.loc['교육대']
all_school
```

```python
all_school = all_school.drop(['전문대','교육대'], axis=0)
```

## 년도별 경기도 학교 수

```python
ggd_school = ggd_school.set_index('교육과정별(1)')
ggd_school = ggd_school.astype(int)
ggd_school
```

```python
ggd_school.loc['대학교'] = ggd_school.loc['전문대학 및 대학'] + ggd_school.loc['대학교'] + ggd_school.loc['교육대학']
ggd_school
```

```python
ggd_school = ggd_school.drop(['전문대학 및 대학','교육대학'], axis=0)
ggd_school
```

```python
ggd_school2 = ggd_school.drop('합계',axis=0)
ggd_school2 = ggd_school2[['2019','2020']]
ggd_school2
```

```python
#컬럼 이름 변경
ggd_school2 = ggd_school2.rename(columns={'2019':'경기도'})
ggd_school2
```

```python
label = ['초등학교','중학교','고등학교','대학교']
```

```python
plt.pie(ggd_school2['경기도'],labels = label,autopct='%.1f%%',explode=(0.05,0,0,0))
plt.title('경기도 학교')
```

```python
# 전국학교수 경기도 학교수 데이터합치고 전국의 2019컬럼을 전국으로 변경
df = pd.concat([all_2019,ggd_school2], axis=1)
df = df.rename(columns={'2019':'전국'})
df
```

```python
df = df[['전국','경기도']]
df=df[['경기도','전국']] # 열 순서 변경
```

```python
sns.set_palette('pastel')
plt.figure(figsize=(70,15))
df.plot(kind = 'bar', stacked=True)  #누적막대그래프
plt.title('전국 / 경기도 학교 수', fontsize=15)
plt.legend(fontsize = 10)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
```
