## 경기도 연령대

```python
age_ggd = pd.read_csv('./미니1차/경기도연령대비율_2019.csv', encoding='euc-kr')
age_ggd
```

```python
# 0번째 인덱스 삭제
age_ggd= age_ggd.drop(0, axis=0)
```

```python
age_ggd
```

```python
# 인덱스 재설정
age_ggd = age_ggd.reset_index(drop=True)
```

```python
# 열이름 변경
age_ggd = age_ggd.rename(columns={'구분별(2)':'시군명'})
#age_ggd[['15-19세','20-29세','30-39세','40-49세','50-59세','60세이상']]= age_ggd[['15-19세','20-29세','30-39세','40-49세','50-59세','60세이상']].astype(int)
```

```python
age_ggd.info()
```

```python
age_ggd
```

```python
## 특정 열의 값끼리 계산해서 새로운 컬럼 추가
# 경기도 구역별 총인구수에 연령대 비율 값 계산해서 해당 연령 수 구하기
age_ggd['15~19세 수'] = round(ggd_si['2019'] * age_ggd['15-19세']*1/100)
age_ggd['20~29세 수'] = round(ggd_si['2019'] * age_ggd['20-29세']*1/100)

# 특정 컬럼 뽑아서 새로운 변수에 저장
age_count = age_ggd[['시군명','15~19세 수','20~29세 수']]

# 시군명 컬럼을 인덱스로 지정
age_count = age_count.set_index('시군명')
age_count
```

```python
# 열의 값을 행기준으로 더해서 새로운 컬럼으로 저장
age_count['합'] = age_count['15~19세 수'] + age_count['20~29세 수']
```

```python
# '합' 컬럼으로 내림차순 정렬
age_under_29 = age_count.sort_values(by='합', ascending=False).head(5)
age_under_29
```

```python
# 15~19세 수 컬럼으로 내림차순 정렬 
age_count.sort_values(by='15~19세 수', ascending=False).head(5)
```

### 경기도 학교

```python
ggd_school = pd.read_csv('./미니1차/경기도총학교수.csv', encoding='euc-kr')
```

```python
# 0번째 인덱스 삭제
ggd_school = ggd_school.drop(0, axis=0)
# 교육과정별(1) 컬럼을 인덱스로 설정 
ggd_school = ggd_school.set_index('교육과정별(1)')
ggd_school
```

```python
# 컬럼 값 int 타입으로 변경
ggd_school = ggd_school.astype(int)

# 교육대학 행 삭제
ggd_school = ggd_school.drop('교육대학', axis=0)
ggd_school
```

```python
# 특정 행의 값끼리 열기준으로 더해서 새로운 행 만들기
ggd_school.loc['대학교'] = ggd_school.loc['전문대학 및 대학'] + ggd_school.loc['대학교']
ggd_school
```

```python
# 필요없는 행 삭제
ggd_school_2019= ggd_school.drop(['전문대학 및 대학'])
ggd_school_2019
```

### 인구 연령대

```python
ggd_si_index = ggd_si.set_index('행정구역(시군구)별')
# 특정 행의 값만 가져오기
ggd_si_index = ggd_si_index.loc[['수원시','고양시','용인시','성남시','부천시']]
ggd_si_index
```

```python
# 15~29세 인구수 데이터에서 합뽑기
age_total = age_count[['합']]
age_total = age_total.loc[['수원시','고양시','용인시','성남시','부천시']]
age_total
```

```python
plt.rc("font", family = "Malgun Gothic", size=10)
plt.figure(figsize=(50,5))
fig, ax1 = plt.subplots()
ax1.bar(ggd_si_index.index, ggd_si_index['2019'],color='violet', label='Demand', alpha=0.7, width=0.7)

plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.title('-', fontsize= 10)

ax2 = ax1.twinx()
ax2.plot(age_total.index, age_total['합'],'-s', color='red', markersize=5, linewidth=3, alpha=0.7)
plt.ylim()
```

```python
# 위 두개의 데이터프레임 합치기
ingu_age = pd.concat([ggd_si_index,age_total['합']], axis=1)
ingu_age
```

```python
sns.set_palette('pastel')
plt.figure(figsize=(70,15))
ingu_age.plot(kind = 'bar')
plt.title('경기도 총 인구수 대 15~29세 인구수', fontsize=15)
plt.legend(fontsize = 10)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
```
