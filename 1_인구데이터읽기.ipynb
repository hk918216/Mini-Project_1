## 경기도로 유입되는 각 지역의 인구수

```python
move = pd.read_csv('./미니1차/전출지_전입지_시도_별_이동자수_20220622110814.csv',encoding = 'euc-kr')
move
```

```python
# '2022.02','2022.03','2022.04' 열 삭제
move = move.drop(['2022.02','2022.03','2022.04'],axis=1)

# 인덱스 재설정
move = move.set_index(['전출지별'])
```

```python
move.head()
```

```python
# 전출지에서 각 경기도로 전입하는 인구수만 찾기
move_ggd = move[move['전입지별']=='경기도']
move_ggd 
```

```python
# [세종특별자치시] 인덱스의 2010,11 열 값 0으로 변경
move_ggd.loc['세종특별자치시',['2010','2011']] = 0
```

```python
# 전입지별 열 삭제
move_ggd = move_ggd.drop('전입지별', axis=1)
```

```python
# 경기도 행 삭제
move_ggd = move_ggd.drop('경기도',axis=0)
```

```python
# int 타입으로 변경
move_ggd = move_ggd.astype(int)
# '전국' 행 삭제
move_ggd= move_ggd.drop('전국',axis=0)
```

```python
move_ggd
```

```python
# [합계]라는 새로운 컬럼 만들고 행기준으로 더해서 추가
move_ggd['합계']=move_ggd.sum(axis=1)
```

```python
move_ggd
```

```python
move_ggd.to_csv('./가공데이터/시도별경기도전입인구.csv')
```
