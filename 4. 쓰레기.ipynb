```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
trash = pd.read_csv('C:/Users/admin/STUDY/Mini-Project_1/미니1차/전국폐기물발생및처리현황.csv', encoding='euc-kr')
trash
```

```python
trash = trash.drop(0, axis=0)
```

```python
trash = trash[['시도','폐기물 종류','2019발생량']]
```

```python
trash # 불필요 행 삭제 후 필요한 컬럼만 불러오기
```

```python
trash = trash[trash['폐기물 종류'] == '합계']
trash # 필요 값만 추출
```

```python
trash = trash.set_index('시도')
trash = trash.drop('폐기물 종류',axis=1)
trash
```

```python
trash = trash.drop('전국',axis=0)
```

```python
# trash= trash.sort_values(by='2019발생량', ascending=False).head()
# trash
```

```python
trash = trash.astype(int)
```

```python
trash = trash.sort_values(by='2019발생량', ascending=False)
```

```python
# labels = ['경기','서울','경남','부산','대구']
color = ['#6c5ce7','#d63031','#81ecec','#00b894','#fdcb6e']
```

```python
plt.rc("font", family = "Malgun Gothic", size=20)
plt.figure(figsize=(70,15))
plt.pie(trash['2019발생량'],labels = trash.index,colors = color, autopct='%.1f%%') #,explode=(0.05,0,0,0,0) 해당위치 값 돌출시키기
plt.title('전국 생할쓰레기 배출량')
# 다섯개의 지역사이에서의 비율을 나타냄
```
